```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
x = np.linspace(0, 100, 100)
pff = -1*x+100
pff2 = -1*(x-10)+100
fig, ax = plt.subplots()

ax.plot(x, pff)
ax.plot(x, pff2)
ax.fill_between(x, pff, color='skyblue')

xmin = min(x.min(), x.min())
xmax = max(x.max(), x.max())
ymin = min(pff.min(), pff2.min())
ymax = max(pff.max(), pff2.max())
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

ax.set_xlabel('Bicycles (arbitrary units)')
ax.set_ylabel('Other Goods (arbitrary units)')
ax.set_title('Production Possibilities Frontier for Bicycles VS Other Goods in the UK')
plt.show()
```

```python
x = np.linspace(0, 100, 10)
pff = -0.8*x+80
pff2 = -1.2*(x+20)+120
fig, ax = plt.subplots()

ax.plot(x, pff, label='comparative advantage in bicycles')
ax.plot(x, pff2, label='comparative advantage in other goods')
#ax.fill_between(x, pff, color='skyblue')

xmin = min(x.min(), x.min(), 0)
xmax = max(x.max(), x.max())
ymin = min(abs(pff.min()), abs(pff2.min()), 0)
ymax = max(pff.max(), pff2.max())
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

ax.legend(loc='best')

ax.set_xlabel('Bicycles (arbitrary units)')
ax.set_ylabel('Other Goods (arbitrary units)')
ax.set_title('Production Possibilities Frontier for Bicycles VS Other Goods in China')
plt.show()
```

```python
local_supply = np.ones(2)*20
import_supply = np.ones(2)*50
intl_price = np.ones(2)*20
dom_price = np.ones(2)*50

fig, ax = plt.subplots()
dom_p, = ax.plot([0,100], dom_price, label='Price Without Trade')
intl_p, = ax.plot([0,100], intl_price, label='Price With Free Trade')
ax.plot(local_supply, [0,max(intl_price)])
ax.plot(import_supply, [0,max(intl_price)])

ax.fill_between([0, max(local_supply), max(local_supply), 0], [0, 0, max(intl_price),max(intl_price)], color='green')
ax.fill_between([max(local_supply), max(import_supply), max(import_supply), max(local_supply)], [0, 0, max(intl_price),max(intl_price)], color='red')
green_patch = mpatches.Patch(color='green', label='Local Supply')
red_patch = mpatches.Patch(color='red', label='Import Supply')

xmin = min(x.min(), x.min(), 0)
xmax = max(x.max(), x.max())
ymin = min(abs(intl_price.min()), abs(dom_price.min()), 0)
ymax = max(intl_price.max(), dom_price.max(),100)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])

ax.legend(loc='best',handles=[dom_p, intl_p, green_patch, red_patch])

ax.set_xlabel('Quantity (arbitrary units)')
ax.set_ylabel('Price (arbitrary units)')
ax.set_title('UK\'s Market for Bicycles, With and Without Trade')
plt.show()
```
